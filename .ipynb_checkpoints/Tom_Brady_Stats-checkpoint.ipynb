```python
import pandas as pd
import sqlite3
import numpy as np
import os
import matplotlib.pyplot as plt

#Reading CSV
QBstats = pd.read_csv('QBStats_all.csv')

#Creating database
QBstats.to_sql('TBstats', sqlite3.connect('nfl_stats.db'), if_exists = "replace")

QBstats.shape
```

__Is there a relationship between Tom Brady's passing yards and the Patriots' game points?__

The following dataset contains all statistics for each NFL quarterback from 1996-2016.

```python
QBstats
```

```python
conn = sqlite3.connect('nfl_stats.db')
pd.read_sql_query('SELECT * FROM TBstats LIMIT 10', conn)
```

Statistics for Tom Brady only. 

```python
pd.read_sql_query('SELECT * FROM TBstats WHERE qb = "Tom Brady"', conn)
```

__Tom Brady's passing yards vs game points__

```python
all_yds_points = pd.read_sql_query('SELECT yds, game_points FROM TBstats WHERE qb = "Tom Brady"', conn)
all_yds_points
```

__Calculating the average amount of points scored when Brady has less than 150 passing yards.

```python
pd.read_sql_query('SELECT qb, AVG(game_points) AS average_pts FROM TBstats WHERE qb = "Tom Brady" AND yds < 150', conn)
```

__Calculating the average amount of points scored when Brady has less 151-300 passing yards.

```python
pd.read_sql_query('SELECT qb, AVG(game_points) AS average_pts FROM TBstats WHERE qb = "Tom Brady" AND yds BETWEEN 151 AND 300', conn)
```

__Calculating the average amount of points scored when Brady has more than 300 passing yards.

```python
pd.read_sql_query('SELECT qb, AVG(game_points) AS average_pts FROM TBstats WHERE qb = "Tom Brady" AND yds > 300', conn)
```

```python
scatter_of_all = all_yds_points.plot(kind = "scatter", x = 'game_points', y = "yds", title = "Tom Brady's Career Yards vs Points", figsize =(8, 6))
```

__There is a positive and linear relationship between the passing yards and game points, as well as a pattern among the passing yardage breakdowns. When Brady's passing yards increase, the avg score does as well.
